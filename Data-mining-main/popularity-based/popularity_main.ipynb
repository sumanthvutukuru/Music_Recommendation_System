```python
import pandas as pd

data = pd.read_table('../data/kaggle_visible_evaluation_triplets.txt', delim_whitespace=True, names=('A', 'B', 'C'))
data
```

```python
msd_songs=pd.read_table('../data/kaggle_songs.txt', delim_whitespace=True, names=('B', 'song_id'))
msd_songs
```

```python
msd_user=pd.read_table('../data/kaggle_users.txt', delim_whitespace=True)
msd_user
```

```python
new_data=pd.merge(data,msd_songs,on='B')
new_data
```

```python
new_data['song_id'].value_counts()
```

```python
value_counts = new_data['song_id'].value_counts()

# converting to df and assigning new names to the columns
df_value_counts = pd.DataFrame(value_counts)
df_value_counts = df_value_counts.reset_index()
df_value_counts.columns = ['song_id', 'count'] # change column names
df_value_counts
```

```python
cnt=0
for indx in df_value_counts.index:
    if(df_value_counts['count'][indx]>2):
        cnt=cnt+df_value_counts['count'][indx]
    else:
        break
cnt
```

```python
df_value_counts['song_id'].head(500).to_csv('popularity.csv', index=False,header=False)
```
